## Exercise - Analyzing Data Science Job Demand

```python
#import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as pt 
# load data 
dataset = load_dataset('csv', data_files='/Users/onurtemel/Downloads/data_jobs.csv')
df = dataset["train"].to_pandas()
# Data Clean up
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df_US = df[df["job_country"] == "United States"].copy()
#all job offers from 2023 we want monthh to be verb such as january
df_US
```

```python
df_US["job_posted_month"] = df_US["job_posted_date"].dt.strftime("%B")
df_US
```

```python
df_US_pivot = df_US.pivot_table(index = "job_posted_month",columns="job_title_short", aggfunc="size")
df_US_pivot
```

```python

df_US_pivot.plot(kind = "line")

```

```python
top_3 = df_US["job_title_short"].value_counts().head(3)
top_3 = top_3.index.to_list()
df_US_pivot[top_3].plot(kind = "line")
pt.xlabel("Months")
pt.ylabel("Number of jobs")
```

```python
df_US_pivot
```

```python
df_US_software_pivot = pd.read_csv("https://lukeb.co/software_csv",index_col = "job_posted_month")
```

```python
df_US_pivot
```

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot,on="job_posted_month")
```

```python
top_5 = (df_US_merged
.sum()
.sort_values(ascending = False)
.head()
.index.to_list())
top_5
```

```python
df_US_merged[top_5].plot(kind = "line")
pt.title("Montly Job Postings for Top Tech in the US")
pt.xlabel("2023")
pt.ylabel("Job Count")
pt.ylim(0,20000)
pt.legend()
pt.show()
```
